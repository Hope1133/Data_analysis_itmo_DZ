## Задача 1
2.8. В ящике 10 красных и 6 синих пуговиц. Какова вероятность того, что две наудачу вынутые пуговицы будут одноцветными?

```python
import random
```

```python
#  Задаем зерно датчика  a, 
#  опция version=2 требует преобразовать а в целое число, если требуется
random.seed(a=42, version=2)

#  Пространство элементарных событий - всего 16 пуговиц, порядка нет
A = ['R']*10 + ['B']*6

delta_ = 1
p_0 = 1
n_exp = 10000

while delta_ >= 0.0001:
    c=0
    for i in range(n_exp):
        a = random.sample(A, 2)
        if (a[0] == a[1]):
            c+=1
        
    p = c/n_exp              # Оцениваем вероятность
    delta_ = abs(p - p_0)        # Считаем изменение вероятности
    p_0 = p                      # Новая оценка вероятности заменяет старую
    n_exp = round(n_exp * 1.1)   # Увеличиваем число экспериментов

print(round(p, 4))
```

Правильный ответ:

## Задача 2
2.17. N человек, среди которых находятся А и Б, случайным образом поставлены в ряд. Какова вероятность того, что между А и Б окажется ровно r человек?

```python
N = 100
r = 30
```

```python
#  Задаем зерно датчика  a, 
#  опция version=2 требует преобразовать а в целое число, если требуется
random.seed(a=42, version=2)

#  Пространство элементарных событий - все порядковые номера от 1 до N
A = list(range(1, N+1))

delta_ = 1
p_0 = 1
n_exp = 10000

while delta_ >= 0.0001:
    c=0
    for i in range(n_exp):
        a = random.sample(A, 2)
        if (abs(a[0] - a[1]) == r):
            c+=1
        
    p = c/n_exp              # Оцениваем вероятность
    delta_ = abs(p - p_0)        # Считаем изменение вероятности
    p_0 = p                      # Новая оценка вероятности заменяет старую
    n_exp = round(n_exp * 1.1)   # Увеличиваем число экспериментов

print(round(p, 4))
```

Правильный ответ:

```python
import scipy.special

print((N-(r+1))*2/ (N*(N-1)))
```

## Задача 3
Стержень длины l наудачу разламывается на две части, после чего б´ольшая из частей опять разламывается надвое в наудачу выбранной точке. Найти вероятность того, что из полученных частей можно составить треугольник.

```python
l = 10
```

```python
#  Задаем зерно датчика  a, 
#  опция version=2 требует преобразовать а в целое число, если требуется
random.seed(a=42, version=2)

delta_ = 1
p_0 = 1
n_exp = 10000

while delta_ >= 0.0001:
    c=0
    for i in range(n_exp):
        first = random.uniform(0, l)
        if first <= l/2:
            # Меньшая часть = first, большая = l - first
            # Второй разрыв на большей части [0, l-first]
            second = random.uniform(0, l - first)
            # Три отрезка: first, second, l-first-second
            l1 = first
            l2 = second
            l3 = l - first - second
        else:
            # Меньшая часть = l-first, большая = first
            # Второй разрыв на большей части [0, first]
            second = random.uniform(0, first)
            # Три отрезка: l-first, second, first-second
            l1 = l - first
            l2 = second
            l3 = first - second

        if (l1+l2 > l3) & (l1+l3 > l2) & (l3+l2 > l1):
            c+=1
        
    p = c/n_exp              # Оцениваем вероятность
    delta_ = abs(p - p_0)        # Считаем изменение вероятности
    p_0 = p                      # Новая оценка вероятности заменяет старую
    n_exp = round(n_exp * 1.1)   # Увеличиваем число экспериментов

print(round(p, 4))
```

Правильный ответ:

1. Первый разрыв:  
Пусть $x \leq l/2$, меньшая часть имеет длину $x$, большая: $l - x$.

2. Второй разрыв:  
Получаются три отрезка: $x$, $y$, $l - x - y$.

3. Условия существования треугольника:
   - $x + y > l - x - y \quad => \quad x + y > l/2$
   - $x + (l - x - y) > y \quad => \quad  y < l/2$
   - $y + (l - x - y) > x \quad => \quad  x < l/2$

   Для $x \in (0, l/2)$ благоприятные значения $y$ удовлетворяют неравенствам:  
   $$
   \frac{l}{2} - x < y < \frac{l}{2}, \quad 0 \leq y \leq l - x.
   $$

4. Вычисление вероятности:  
   Совместная плотность распределения $(x, y)$:
   $$
   f(x, y) = \frac{1}{l} \cdot \frac{1}{l - x}, \quad 0 \leq x \leq \frac{l}{2}, \; 0 \leq y \leq l - x.
   $$
   Учитывая симметрию случая:
   $$
   P = 2 \int_{0}^{l/2} \int_{l/2 - x}^{l/2} \frac{1}{l(l - x)} \, dy \, dx = 2 \int_{0}^{l/2} \frac{x}{l(l - x)} \, dx
   $$

   $$
   \int_{0}^{l/2} \frac{x}{l - x} \, dx = l \ln 2 - \frac{l}{2}
   $$
   $$
   P = 2 \left( \ln 2 - \frac{1}{2} \right) = 2 \ln 2 - 1.
   $$
